```python
import pandas as pd
filename = 'u.user.txt'
data = pd.read_csv(filename, delimiter='|', header = 0)
print(data)
```

```python
data.info()
```

```python
data.groupby(['occupation']).mean()
```

```python
q3 = data.groupby('occupation')['gender'].value_counts(normalize=True).reset_index(name='perc').sort_values(by='perc', ascending=False)
print(q3)
```

```python
q4 = data.groupby('occupation').agg({'age': ['min', 'max']})
print(q4)
```

```python
q5 = data.groupby(['occupation', 'gender']).agg({'age': ['mean']})
print(q5)
```

```python
q6 = data.groupby('occupation')['gender'].value_counts(normalize=True).reset_index(name='perc')
print(q6)
```

```python
import csv
import matplotlib.pyplot as plt
import pandas as pd


data = pd.read_csv('actor_kill_counts.csv')
df = pd.DataFrame(data)
df = df.sort_values(by=['Count'])
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])


plt.barh(X, Y, color='purple')
plt.title("Actor Kill Count")
plt.ylabel("Actor")
plt.xlabel("Kill count")



plt.show()
```

```python
import csv
from matplotlib.pyplot import pie, axis, show

%matplotlib inline
emperors = pd.read_csv('roman-emperor-reigns.csv')
emperors = emperors.loc[(emperors["Cause_of_Death"] == "Assassinated")]
emperors = emperors.sort_values(by=['Length_of_Reign'])

labels = [x[1] for x in emperors.columns]



sums = emperors.groupby(emperors["Emperor"])["Length_of_Reign"].sum()
axis('equal');
pie(sums, labels=sums.index, autopct='%1.1f%%');

show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = df['Total Arcade Revenue (billions)'], y = df['Computer Science Doctorates Awarded (US)'])
plt.title("Relation Between TAR and CSDA (US)")
plt.xlabel("Total Arcade Revenue (billions)")
plt.ylabel("Computer Science Doctorates Awarded (US)")

plt.show()
```
